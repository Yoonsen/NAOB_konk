### Sammenlign ulike naob-korpusfiler


```python

import pandas as pd


naob_csv = "naob.csv"

naob_excel = "naob.xlsx"
naob_excel2 = "naob2.xlsx"

naob = pd.read_csv(naob_csv, index_col=0)

naobx = pd.read_excel(naob_excel, index_col=0)
naobx["corpus"] = "naob_excel"

naobx2 = pd.read_excel(naob_excel2, index_col=0)
naobx2["corpus"] = "naob_excel2"
```

```python
naob[['urn', 'corpus']].drop_duplicates("urn", keep="first")
```

```python

# Check that all URNS unmarked with the time of the last update are duplicated by URNs that are marked
naob.loc[~naob.loc[naob.corpus.isna(), "urn"].isin(naob.loc[naob.corpus.notna(), "urn"]), :]
```

```python

# We can safely drop the unmarked URNs
naob.dropna(subset=["corpus"])
```

```python
# Drop duplicated urns from the excel file 
first = naobx[['urn', 'corpus', "title"]].drop_duplicates("urn")
second = naobx2[['urn', 'corpus', "title"]].drop_duplicates("urn")
```

```python


# Check that all URNs in the first file are also in the second file
all(first.urn.isin(second.urn))
```

```python

# And check that all urns from the excel file are in the original csv file 
all(first.urn.isin(naob.urn))
```
